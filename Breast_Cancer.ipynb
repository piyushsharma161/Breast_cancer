```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
y_raw = pd.get_dummies(y, drop_first=True)
y_1 = y_raw.iloc[:,0]
```

```python
#with raw data
X_train, X_test, y_train, y_test = train_test_split(X, y_1, test_size=0.3, random_state=42)
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rfp = rf.predict(X_test)
print('accuracy is: ',accuracy_score(y_test, rfp) )
print('f1_score is: ', f1_score(y_test, rfp))
print('\n')
print('confusion matrixis : \n')
print(confusion_matrix(y_test, rfp))
print('\n')
print('classification score : ')
print(classification_report(y_test, rfp))
sns.heatmap(confusion_matrix(y_test, rfp), annot=True, fmt='d')
```

# Univariate feature selection and random forest classification

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# only with 16 features(that we extracted manually)
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size=0.3, random_state=42)
```

```python
select_feature = SelectKBest(chi2, k=5).fit(X_train, y_train)
```

```python
print('score list: \n', select_feature.scores_)
print('columns list : \n', X_train.columns)
```

```python
d = {'feature':list(X_train.columns), 'score':select_feature.scores_}
df = pd.DataFrame(d)
df.sort_values('score', ascending=False)
```

# now transform both traing and test using the above feature selction algo(select_feature)

```python
X_train_2 = select_feature.transform(X_train)
y_test_2 = select_feature.transform(X_test)                    
```

```python
rf2 = RandomForestClassifier(random_state=42)
rf2 = rf2.fit(X_train_2, y_train)
rfp2 = rf2.predict(y_test_2)
print('accuracy is: ',accuracy_score(y_test, rfp2) )
print('f1_score is: ', f1_score(y_test, rfp2))
print('\n')
print('confusion matrixis : \n')
print(confusion_matrix(y_test, rfp2))
print('\n')
print('classification score : ')
print(classification_report(y_test, rfp2))
sns.heatmap(confusion_matrix(y_test, rfp2), annot=True, fmt='d')
```

# Recursive feature elimination (RFE) with random forest

```python
from sklearn.feature_selection import RFE
# Create the RFE object and rank each pixel
rfcm1 = RandomForestClassifier(random_state=101)
rf3 = RFE(rfcm1, n_features_to_select=5, step=1)
rf3 = rf3.fit(X_train, y_train)
```

```python
X_train.columns[rf3.support_]
```

```python
rfp3 = rf3.predict(X_test)
print('accuracy is: ',accuracy_score(y_test, rfp3) )
print('f1_score is: ', f1_score(y_test, rfp3))
print('\n')
print('confusion matrixis : \n')
print(confusion_matrix(y_test, rfp3))
print('\n')
print('classification score : ')
print(classification_report(y_test, rfp3))
sns.heatmap(confusion_matrix(y_test, rfp3), annot=True, fmt='d')
```

# Recursive feature elimination with cross validation and random forest classification

```python
from sklearn.feature_selection import RFECV
rf4 = RandomForestClassifier(random_state=101)
rfcv = RFECV(rf4, cv=5, scoring='accuracy', step=1)
rfcv = rfcv.fit(X_train, y_train)
```

```python
print("optimal number of features: \n", rfcv.n_features_)
print('best features are: \n', X_train.columns[rfcv.support_])
```

```python
# Plot number of features VS. cross-validation scores
plt.figure(figsize=(11,7))
plt.plot(range(1, len(rfcv.grid_scores_)+1), rfcv.grid_scores_)
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score of number of selected features")
```

```python
rfcv.grid_scores_
```
